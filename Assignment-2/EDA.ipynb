###  Let's Transform the flipkart and amazon dataset into something useful throught which we can recommend product to user from the dataset 

```python
import pandas as pd 
flipkart = pd.read_csv('/content/drive/MyDrive/Shack Labs Assignment/Assignment-2/flipkart_com-ecommerce_sample.csv')
amazon = pd.read_csv("/content/drive/MyDrive/Shack Labs Assignment/Assignment-2/amz_com-ecommerce_sample.csv",encoding='latin-1')
```

```python
flipkart = flipkart.replace(np.nan,'')
amazon = amazon.replace(np.nan,'')
```

```python
flipkart = flipkart.drop_duplicates()
amazon = amazon.drop_duplicates()
```

```python
flipkart.shape
```

```python
flipkart_data = flipkart[['product_name','product_category_tree','retail_price','discounted_price','description','brand','product_specifications']]
```

```python
amazon_data = amazon[['product_name','product_category_tree','retail_price','discounted_price','description','brand','product_specifications']]
```

```python
def imputer(value):
  return ' '.join(value[2:-2].split(' >> '))
flipkart_data['product_category_tree'] = flipkart_data['product_category_tree'].apply(imputer)
amazon_data['product_category_tree'] = amazon_data['product_category_tree'].apply(imputer)
```

```python
import numpy as np
flipkart_data = flipkart_data.replace(np.nan,'')
amazon_data = amazon_data.replace(np.nan,'')

def remove_punc(string):
  if  string == np.nan:
    return ""
  string = string.replace('key',' ')
  string = string.replace('value',' ')
  punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~='''
  for c in string :
    if c in punc :
      string = string.replace(c, "")
  return string 
  
flipkart_data['product_specifications'] = flipkart_data['product_specifications'].apply(remove_punc)
amazon_data['product_specifications'] = amazon_data['product_specifications'].apply(remove_punc)
```

```python
amazon_data['product_info']=amazon_data['product_name'] +' '+ amazon_data['brand']+' ' + amazon_data['description'] + ' ' + amazon_data['product_category_tree'] + ' ' + amazon_data['product_specifications']
AZ_products = amazon_data.drop(['product_category_tree','description','brand','product_specifications'],axis=1)
```

```python
flipkart_data['product_info']=flipkart_data['product_name'] +' '+ flipkart_data['brand']+' ' + flipkart_data['description'] + ' ' + flipkart_data['product_category_tree'] + ' ' + flipkart_data['product_specifications']
FK_products = flipkart_data.drop(['product_category_tree','description','brand','product_specifications'],axis=1)
```

```python
AZ_products
```

```python
FK_products
```

```python
FK_products = FK_products.drop_duplicates(subset=['product_name'])
AZ_products = AZ_products.drop_duplicates(subset=['product_name'])
```

```python
FK_products.shape
```

```python
AZ_products.shape
```

```python
FK_products = FK_products.reset_index(drop=True)
AZ_products = AZ_products.reset_index(drop=True)
```

```python
FK_products.to_csv('transformed Flipkart data.csv')
AZ_products.to_csv('transformed Amazon Data.csv')
```
